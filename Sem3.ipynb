```python
import opendatasets as od
import pandas as pd
from functools import reduce
from pathlib import Path
import csv

# Загрузка датасета, если он еще не скачан
dataset_path = Path('cardataset', 'data.csv')
if not dataset_path.is_file():
    od.download('https://www.kaggle.com/datasets/CooperUnion/cardataset')

# Функции Map и Reduce
def map_function(row):
    car_type = row['Make']
    price = float(row['MSRP'])
    return (car_type, [price])

def reduce_function(accumulated_data, new_data):
    car_type, price = new_data
    if car_type in accumulated_data:
        accumulated_data[car_type].extend(price)
    else:
        accumulated_data[car_type] = price
    return accumulated_data

# Чтение и обработка данных
with open(dataset_path, "r", encoding='utf8') as file:
    reader = csv.DictReader(file)
    m = map(map_function, reader)
    result = reduce(reduce_function, map(map_function, reader), {})
    print(result)

# Вычисление средней цены
average_prices = {car_type: sum(prices) / len(prices) for car_type, prices in result.items() if prices}

# Вывод результатов
#for car_type, avg_price in average_prices.items():
 #   print(f"{car_type}: {avg_price}")
```

```python
# Подготовка данных с учетом требования о двух знаках после запятой
data_with_precision = {
    "Brand": list(average_prices.keys()),
    "avg_price": [round(price, 2) for price in average_prices.values()]
}

# Создание DataFrame с обновленными данными
df_with_precision = pd.DataFrame(data_with_precision)
df_with_precision.head()
```
